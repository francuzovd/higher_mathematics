### Задание 2.
1. Докажите, что при ортогональном преобразовании сохраняется расстояние между точками.

##### Решение
Пусть даны 2 точки A (x1, y1) и B (X2, y2). Расстояние между ними равно:
$$AB = {\sqrt{(x1 - x2)^2 + (y1 - y2)^2}}$$

Если мы переведём точки A и B ортогональным преобразованием в точки A1 и B1 соответственно, то:
$$x1 - x2 = (x1 - a) * cos(\alpha) + (y1 - b) * sin(\alpha) - (x2 - a) * cos(\alpha) + (y2 - b) * sin(\alpha)$$
$$x1 - x2 = (x1 - x2) * cos(\alpha) + (y1 - y2) * sin(\alpha) $$

$$y1 - y2 = -(x1 - a) * sin(\alpha) + (y1 - b) * cos(\alpha) + (x2 - a) * sin(\alpha) + (y2 - b) * cos(\alpha) $$
$$y1 - y2 = -(x1 - x2) *  sin(\alpha) + (y1 - y2) * cos(\alpha)$$

В этом случае расстояние между этими точками будет равно
$$A_{1}B_{1} = {\sqrt{( (x1 - x2) * cos(\alpha) + (y1 - y2) * sin(\alpha))^2 + (-(x1 - x2) *  sin(\alpha) + (y1 - y2) * cos(\alpha))^2}}$$

Раскрыв скобки мы получим
$$A_{1}B_{1} = \sqrt{(x1 - x2)^2 + (y1 - y2)^2}$$

Т.о. $AB = A_{1}B_{1}$, что доказывает сохранение расстояния при ортогональном преобразовании.

### Задание 4.
1. Решите систему уравнений:

y = x2 – 1

exp(x) + x∙(1 – y) = 1

```python
x = np.linspace(-5, 5, 100)
f1 = x **2 - 1
f2 = 1 - ((1 - np.exp(x)) / x)

plt.plot(x, f1)
plt.plot(x, f2)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x **2 + 1, np.exp(x) + x * (1 - y) - 1)
    
fsolve(equations, (-2, 2)), fsolve(equations, (2, 2)), fsolve(equations, (4, 2))
```
